```python
import pandas as pd
import numpy as np

# create function to return distances of vectors

def get_distance(x, y, px_cm):

    d = np.cumsum(np.sqrt( np.ediff1d(x)**2 + np.ediff1d(y)**2 ))
    d = d[-1]
    d = d/px_cm

    return d

# create empty lists to store distances

m_distances = []
d_distances = []
p_distances = []

m1_distances = []
d1_distances = []
p1_distances = []

m2_distances = []
d2_distances = []
p2_distances = []

m3_distances = []
d3_distances = []
p3_distances = []

ma_distances = []
da_distances = []
pa_distances = []

mb_distances = []
db_distances = []
pb_distances = []

mc_distances = []
dc_distances = []
pc_distances = []

md_distances = []
dd_distances = []
pd_distances = []

ds_distances = []
ps_distances = []

variety = []
leaf = []
image = []

# read in dataframe with leaf ID information

df = pd.read_table('Desktop/0_visual_check/ampelometry_id_key.txt', delim_whitespace=True, encoding="latin1")

# iterate over each row of dataframe to calculate distances for each leaf

for index, row in df.iterrows():
    
    # get leaf to retrieve data from
    
    Variety = row["variety"]
    Leaf = row["value"]
    Image = row["image"]

    cm_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_cm.txt"

    m_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_m.txt"
    d_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_d.txt"
    p_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_p.txt"

    m1_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_m1.txt"
    d1_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_d1.txt"
    p1_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_p1.txt"

    m2_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_m2.txt"
    d2_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_d2.txt"
    p2_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_p2.txt"

    m3_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_m3.txt"
    d3_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_d3.txt"
    p3_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_p3.txt"

    ma_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_ma.txt"
    da_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_da.txt"
    pa_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_pa.txt"

    mb_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_mb.txt"
    db_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_db.txt"
    pb_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_pb.txt"

    mc_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_mc.txt"
    dc_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_dc.txt"
    pc_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_pc.txt"

    md_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_md.txt"
    dd_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_dd.txt"
    pd_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_pd.txt"

    ds_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_ds.txt"
    ps_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_ps.txt"

    cm = np.loadtxt(cm_file)

    m = np.loadtxt(m_file)
    d = np.loadtxt(d_file)
    p = np.loadtxt(p_file)

    m1 = np.loadtxt(m1_file)
    d1 = np.loadtxt(d1_file)
    p1 = np.loadtxt(p1_file)

    m2 = np.loadtxt(m2_file)
    d2 = np.loadtxt(d2_file)
    p2 = np.loadtxt(p2_file)

    m3 = np.loadtxt(m3_file)
    d3 = np.loadtxt(d3_file)
    p3 = np.loadtxt(p3_file)

    ma = np.loadtxt(ma_file)
    da = np.loadtxt(da_file)
    pa = np.loadtxt(pa_file)

    mb = np.loadtxt(mb_file)
    db = np.loadtxt(db_file)
    pb = np.loadtxt(pb_file)

    mc = np.loadtxt(mc_file)
    dc = np.loadtxt(dc_file)
    pc = np.loadtxt(pc_file)

    md = np.loadtxt(md_file)
    dd = np.loadtxt(dd_file)
    pd = np.loadtxt(pd_file)

    ds = np.loadtxt(ds_file)
    ps = np.loadtxt(ps_file)
    
    # calculate distances of each vector and append to lists

    px_cm = np.cumsum(np.sqrt( np.ediff1d(cm[:,0])**2 + np.ediff1d(cm[:,1])**2 ))
    px_cm = px_cm[-1]

    m_distances.append(get_distance(m[:,0], m[:,1], px_cm))
    d_distances.append(get_distance(d[:,0], d[:,1], px_cm))
    p_distances.append(get_distance(p[:,0], p[:,1], px_cm))

    m1_distances.append(get_distance(m1[:,0], m1[:,1], px_cm))
    d1_distances.append(get_distance(d1[:,0], d1[:,1], px_cm))
    p1_distances.append(get_distance(p1[:,0], p1[:,1], px_cm))

    m2_distances.append(get_distance(m2[:,0], m2[:,1], px_cm))
    d2_distances.append(get_distance(d2[:,0], d2[:,1], px_cm))
    p2_distances.append(get_distance(p2[:,0], p2[:,1], px_cm))

    m3_distances.append(get_distance(m3[:,0], m3[:,1], px_cm))
    d3_distances.append(get_distance(d3[:,0], d3[:,1], px_cm))
    p3_distances.append(get_distance(p3[:,0], p3[:,1], px_cm))

    ma_distances.append(get_distance(ma[:,0], ma[:,1], px_cm))
    da_distances.append(get_distance(da[:,0], da[:,1], px_cm))
    pa_distances.append(get_distance(pa[:,0], pa[:,1], px_cm))

    mb_distances.append(get_distance(mb[:,0], mb[:,1], px_cm))
    db_distances.append(get_distance(db[:,0], db[:,1], px_cm))
    pb_distances.append(get_distance(pb[:,0], pb[:,1], px_cm))

    mc_distances.append(get_distance(mc[:,0], mc[:,1], px_cm))
    dc_distances.append(get_distance(dc[:,0], dc[:,1], px_cm))
    pc_distances.append(get_distance(pc[:,0], pc[:,1], px_cm))

    md_distances.append(get_distance(md[:,0], md[:,1], px_cm))
    dd_distances.append(get_distance(dd[:,0], dd[:,1], px_cm))
    pd_distances.append(get_distance(pd[:,0], pd[:,1], px_cm))

    ds_distances.append(get_distance(ds[:,0], ds[:,1], px_cm))
    ps_distances.append(get_distance(ps[:,0], ps[:,1], px_cm))
    
    variety.append(Variety)
    leaf.append(Leaf)
    image.append(Image)
```

```python
# caculate proportional numbers of points for each vector given their distances

total_points = 6000

m_pts = round((m_sum / sum_sum)*total_points, 0)
d_pts = round((d_sum / sum_sum)*total_points, 0)
p_pts = round((p_sum / sum_sum)*total_points, 0)

m1_pts = round((m1_sum / sum_sum)*total_points, 0)
d1_pts = round((d1_sum / sum_sum)*total_points, 0)
p1_pts = round((p1_sum / sum_sum)*total_points, 0)

m2_pts = round((m2_sum / sum_sum)*total_points, 0)
d2_pts = round((d2_sum / sum_sum)*total_points, 0)
p2_pts = round((p2_sum / sum_sum)*total_points, 0)

m3_pts = round((m3_sum / sum_sum)*total_points, 0)
d3_pts = round((d3_sum / sum_sum)*total_points, 0)
p3_pts = round((p3_sum / sum_sum)*total_points, 0)

ma_pts = round((ma_sum / sum_sum)*total_points, 0)
da_pts = round((da_sum / sum_sum)*total_points, 0)
pa_pts = round((pa_sum / sum_sum)*total_points, 0)

mb_pts = round((mb_sum / sum_sum)*total_points, 0)
db_pts = round((db_sum / sum_sum)*total_points, 0)
pb_pts = round((pb_sum / sum_sum)*total_points, 0)

mc_pts = round((mc_sum / sum_sum)*total_points, 0)
dc_pts = round((dc_sum / sum_sum)*total_points, 0)
pc_pts = round((pc_sum / sum_sum)*total_points, 0)

md_pts = round((md_sum / sum_sum)*total_points, 0)
dd_pts = round((dd_sum / sum_sum)*total_points, 0)
pd_pts = round((pd_sum / sum_sum)*total_points, 0)

ds_pts = round((ds_sum / sum_sum)*total_points, 0)
ps_pts = round((ps_sum / sum_sum)*total_points, 0)

points = [m_pts, d_pts, p_pts, m1_pts, d1_pts, p1_pts, m2_pts, d2_pts, p2_pts, m3_pts, d3_pts, p3_pts, ma_pts, da_pts, pa_pts, mb_pts, db_pts, pb_pts, mc_pts, dc_pts, pc_pts, md_pts, dd_pts, pd_pts, ds_pts, ps_pts]

labels = ["m", "d", "p", "m1", "d1", "p1", "m2", "d2", "p2", "m3", "d3", "p3", "ma", "da", "pa", "mb", "db", "pb", "mc", "dc", "pc", "md", "dd", "pd", "ds", "ps"]

import pandas as pandas

points_df = pandas.DataFrame(list(zip(labels, points)), columns = ['segment', 'number'])

points_df.head()

points_df.to_csv('Desktop/1_interpolation/output_number_of_points.txt', sep='\t')
```

```python
# function to retrieve iterpolated points

from scipy.interpolate import interp1d

def interpolation(x, y, number):

    distance = np.cumsum(np.sqrt( np.ediff1d(x, to_begin=0)**2 + np.ediff1d(y, to_begin=0)**2 ))
    distance = distance/distance[-1]

    fx, fy = interp1d( distance, x ), interp1d( distance, y )

    alpha = np.linspace(0, 1, number)
    x_regular, y_regular = fx(alpha), fy(alpha)
    
    return x_regular, y_regular
```

```python
# create a list of lists for each leaf and lists for other ID information

overall_list = []
variety_list = []
image_list = []
leaf_list = []

for index, row in df.iterrows():
    
    # get leaf to retrieve data from
    
    Variety = row["variety"]
    Leaf = row["value"]
    Image = row["image"]

    cm_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_cm.txt"

    m_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_m.txt"
    d_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_d.txt"
    p_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_p.txt"

    m1_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_m1.txt"
    d1_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_d1.txt"
    p1_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_p1.txt"

    m2_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_m2.txt"
    d2_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_d2.txt"
    p2_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_p2.txt"

    m3_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_m3.txt"
    d3_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_d3.txt"
    p3_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_p3.txt"

    ma_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_ma.txt"
    da_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_da.txt"
    pa_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_pa.txt"

    mb_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_mb.txt"
    db_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_db.txt"
    pb_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_pb.txt"

    mc_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_mc.txt"
    dc_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_dc.txt"
    pc_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_pc.txt"

    md_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_md.txt"
    dd_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_dd.txt"
    pd_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_pd.txt"

    ds_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_ds.txt"
    ps_file = "Desktop/0_visual_check/ampelometry_data/" + Leaf + "_ps.txt"

    cm = np.loadtxt(cm_file)

    m = np.loadtxt(m_file)
    d = np.loadtxt(d_file)
    p = np.loadtxt(p_file)

    m1 = np.loadtxt(m1_file)
    d1 = np.loadtxt(d1_file)
    p1 = np.loadtxt(p1_file)

    m2 = np.loadtxt(m2_file)
    d2 = np.loadtxt(d2_file)
    p2 = np.loadtxt(p2_file)

    m3 = np.loadtxt(m3_file)
    d3 = np.loadtxt(d3_file)
    p3 = np.loadtxt(p3_file)

    ma = np.loadtxt(ma_file)
    da = np.loadtxt(da_file)
    pa = np.loadtxt(pa_file)

    mb = np.loadtxt(mb_file)
    db = np.loadtxt(db_file)
    pb = np.loadtxt(pb_file)

    mc = np.loadtxt(mc_file)
    dc = np.loadtxt(dc_file)
    pc = np.loadtxt(pc_file)

    md = np.loadtxt(md_file)
    dd = np.loadtxt(dd_file)
    pd = np.loadtxt(pd_file)

    ds = np.loadtxt(ds_file)
    ps = np.loadtxt(ps_file)
    
    mx, my = interpolation(m[:,0], m[:,1], m_pts)
    dx, dy = interpolation(d[:,0], d[:,1], d_pts)
    px, py = interpolation(p[:,0], p[:,1], p_pts)

    m1x, m1y = interpolation(m1[:,0], m1[:,1], m1_pts)
    d1x, d1y = interpolation(d1[:,0], d1[:,1], d1_pts)
    p1x, p1y = interpolation(p1[:,0], p1[:,1], p1_pts)

    m2x, m2y = interpolation(m2[:,0], m2[:,1], m2_pts)
    d2x, d2y = interpolation(d2[:,0], d2[:,1], d2_pts)
    p2x, p2y = interpolation(p2[:,0], p2[:,1], p2_pts)

    m3x, m3y = interpolation(m3[:,0], m3[:,1], m3_pts)
    d3x, d3y = interpolation(d3[:,0], d3[:,1], d3_pts)
    p3x, p3y = interpolation(p3[:,0], p3[:,1], p3_pts)

    max, may = interpolation(ma[:,0], ma[:,1], ma_pts)
    dax, day = interpolation(da[:,0], da[:,1], da_pts)
    pax, pay = interpolation(pa[:,0], pa[:,1], pa_pts)

    mbx, mby = interpolation(mb[:,0], mb[:,1], mb_pts)
    dbx, dby = interpolation(db[:,0], db[:,1], db_pts)
    pbx, pby = interpolation(pb[:,0], pb[:,1], pb_pts)

    mcx, mcy = interpolation(mc[:,0], mc[:,1], mc_pts)
    dcx, dcy = interpolation(dc[:,0], dc[:,1], dc_pts)
    pcx, pcy = interpolation(pc[:,0], pc[:,1], pc_pts)

    mdx, mdy = interpolation(md[:,0], md[:,1], md_pts)
    ddx, ddy = interpolation(dd[:,0], dd[:,1], dd_pts)
    pdx, pdy = interpolation(pd[:,0], pd[:,1], pd_pts)

    dsx, dsy = interpolation(ds[:,0], ds[:,1], ds_pts)
    psx, psy = interpolation(ps[:,0], ps[:,1], ps_pts)

    mvals = list(zip(mx, my))
    mvals = list(sum(mvals, ()))
    pvals = list(zip(px, py))
    pvals = list(sum(pvals, ()))
    dvals = list(zip(dx, dy))
    dvals = list(sum(dvals, ()))

    m1vals = list(zip(m1x, m1y))
    m1vals = list(sum(m1vals, ()))
    p1vals = list(zip(p1x, p1y))
    p1vals = list(sum(p1vals, ()))
    d1vals = list(zip(d1x, d1y))
    d1vals = list(sum(d1vals, ()))

    m2vals = list(zip(m2x, m2y))
    m2vals = list(sum(m2vals, ()))
    p2vals = list(zip(p2x, p2y))
    p2vals = list(sum(p2vals, ()))
    d2vals = list(zip(d2x, d2y))
    d2vals = list(sum(d2vals, ()))

    m3vals = list(zip(m3x, m3y))
    m3vals = list(sum(m3vals, ()))
    p3vals = list(zip(p3x, p3y))
    p3vals = list(sum(p3vals, ()))
    d3vals = list(zip(d3x, d3y))
    d3vals = list(sum(d3vals, ()))

    mavals = list(zip(max, may))
    mavals = list(sum(mavals, ()))
    pavals = list(zip(pax, pay))
    pavals = list(sum(pavals, ()))
    davals = list(zip(dax, day))
    davals = list(sum(davals, ()))

    mbvals = list(zip(mbx, mby))
    mbvals = list(sum(mbvals, ()))
    pbvals = list(zip(pbx, pby))
    pbvals = list(sum(pbvals, ()))
    dbvals = list(zip(dbx, dby))
    dbvals = list(sum(dbvals, ()))

    mcvals = list(zip(mcx, mcy))
    mcvals = list(sum(mcvals, ()))
    pcvals = list(zip(pcx, pcy))
    pcvals = list(sum(pcvals, ()))
    dcvals = list(zip(dcx, dcy))
    dcvals = list(sum(dcvals, ()))

    mdvals = list(zip(mdx, mdy))
    mdvals = list(sum(mdvals, ()))
    pdvals = list(zip(pdx, pdy))
    pdvals = list(sum(pdvals, ()))
    ddvals = list(zip(ddx, ddy))
    ddvals = list(sum(ddvals, ()))

    psvals = list(zip(psx, psy))
    psvals = list(sum(psvals, ()))
    dsvals = list(zip(dsx, dsy))
    dsvals = list(sum(dsvals, ()))
    
    allvals = mvals + pvals + dvals + m1vals + p1vals + d1vals + m2vals + p2vals + d2vals + m3vals + p3vals + d3vals + mavals + pavals + davals + mbvals + pbvals + dbvals + mcvals + pcvals + dcvals + mdvals + pdvals + ddvals + psvals + dsvals

    overall_list.append(allvals)
    variety_list.append(Variety)
    image_list.append(Image)
    leaf_list.append(Leaf)
```
